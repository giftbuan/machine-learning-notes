```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('GroceryData.csv')
df.head()
```

```python
plt.bar(df['Name'], df['QtySold'])
plt.title('Quantity Sold per Item')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
all_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_summary = df.groupby('DaysSold')['QtySold'].sum()
day_summary = day_summary.reindex(all_days)

plt.figure(figsize=(10, 6))
plt.bar(day_summary.index, day_summary.values)

plt.title('Total Quantity Sold per Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Quantity Sold')

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
#df[df['DaysSold']=='Monday'] this is just to see the contents of the table

#Create a new variable containing only Monday rows
monday_df = df[df['DaysSold'] == 'Monday']

#This is to plot from the monday df
plt.bar(monday_df['Name'], monday_df['QtySold'])

#labels for the table/chart
plt.title('Monday Sales')
plt.xlabel('Item Name')
plt.ylabel('Qunatity Sold')

plt.show()
```

```python
#define the calendar order so it won't use default - alphabetical
all_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

for day in all_days:
    current_day_df = df[df['DaysSold'] == day]
    if not current_day_df.empty:
        plt.bar(current_day_df['Name'], current_day_df['QtySold'])

        plt.title(f'{day} Sales')
        plt.xlabel('Item Name')
        plt.ylabel('Quantity Sold')

        plt.show()
```
