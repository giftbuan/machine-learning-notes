```python
import pandas as pd
from matplotlib import pyplot as plt

# 1. Load data
df = pd.read_csv('2025-05-31-data.csv')

# 2. Match variables to the exact CSV headers seen in your first image
view_count = df['view_count'] # Match: view_count
Likes = df['likes']          # Match: likes (lowercase in your CSV)
ratio = df['ratio']          # Match: ratio

# 3. Set plot style (Using a standard style to avoid the FileNotFoundError)
plt.style.use('ggplot') 
plt.xscale('log')
plt.yscale('log')

# 4. Create scatter plot (Fixed 'Like' vs 'Likes' variable mismatch)
plt.scatter(view_count, Likes, c=ratio, cmap='summer', 
            edgecolors='black', linewidths=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Ratio of Likes and Dislikes')

plt.xlabel('View Count')
plt.ylabel('Likes')

plt.tight_layout()
plt.show()
```
